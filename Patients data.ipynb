```python
import pandas as pd
import numpy as np
```

```python
data={ 'Patient_ID':[101,102,103,104,105],
      'Name':['Arun','Raji','Priya','Kavin','Ramya'],
       'Age':[27,35,None,54,42],
       'Gender':['Male',None,'Female','Male','Female'],
       'Disease' :['Fever','Cough',None,'Anemia','Tyroid'],
      'Stable':['Yes','No','Yes','No',None]
       }
```

```python
import pandas as pd
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
```

```python
df
```

```python
df['Disease'].fillna(df['Disease'].mode()[0],inplace=True)
```

```python
df
```

```python
df['Stable'].fillna(df['Stable'].mode()[0],inplace=True)
```

```python
df
```
