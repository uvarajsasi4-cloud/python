# **Sonar[rock vs mine]LogisticRegression**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/content/sonar.all-data.csv',header=None)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE=LabelEncoder()
```

```python
data[60]=LE.fit_transform(data[60])
```

```python
data
```

```python
X=data.drop(columns=60,axis=1)
Y=data[60]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on data :",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
```

```python
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on data:",testing_data_accuracy)
```
