```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

```

```python
salary_data={
    "Years_Exp":[1,2,3,4,5,6,7,8,9],
    "Salary":[25000,30000,45000,50000,60000,65000,70000,80000,90000]
}
df=pd.DataFrame(salary_data)
```

```python
df
```

```python
x=df[["Years_Exp"]]
y=df["Salary"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("Mean squared error:",mse)
print("R2 score:",r2)
```

```python
new_years_exp = pd.DataFrame({"Years_Exp": [12]})
predicted_Salary = model.predict(new_years_exp)

print("Predicted Years_Exp:", predicted_Salary[0])
```
